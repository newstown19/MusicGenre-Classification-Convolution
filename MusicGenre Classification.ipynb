```python
## Extracting Features from disk
feature = []
for root , dirs , files in os.walk(r'C:\Users\ROHAN SINGH\PROJECTS\MusicGenre\genres'):
       
    for file in files:
        if file.endswith('wav'):
            label = os.path.basename(root)
            path = root
            path_exten = os.path.join(path,file)
            print(path_exten)
            mfcc = feature_extraction(path_exten)
            
            feature.append([mfcc , label])
            
            featuresdf = pd.DataFrame(feature , columns=['feature','class_label'])
            
print('Finished feature extraction from ', len(featuresdf), ' files')            
            
    
```

```python
def feature_extraction(filename):
    max_pad_len = 1320
   
    try:
        sample , sr = librosa.load(filename , res_type='kaiser_fast')
        mfcc = librosa.feature.mfcc(y = sample , sr = sr , n_mfcc=70)
        pad_len = max_pad_len - mfcc.shape[1]
        mfs = np.pad(mfcc , pad_width=((0,0), (0,pad_len)) , mode = 'constant')
        
    except Exception as e :
        print("The file could not be oopened as " , e)
        
        
    return mfs    
```

```python
X = np.asarray(featuresdf.feature.tolist())
Y = np.asarray(featuresdf.class_label.tolist())
```

```python
X[8].shape
```

```python
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
# Convert features and corresponding classification labels into numpy arrays


# Encode the classification labels
le = LabelEncoder()
yy = to_categorical(le.fit_transform(Y)) 

# split the dataset 
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(X, yy, test_size=0.2, random_state = 42 , shuffle = True)
```

```python
x_train = np.expand_dims(x_train , axis = -1)
x_test = np.expand_dims(x_test , axis = -1)
print(x_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense , Dropout , Conv2D , BatchNormalization , Activation , Flatten , MaxPool2D, GlobalAveragePooling2D , AveragePooling2D, LeakyReLU , MaxPooling2D
from keras.optimizers import Adam
from keras.utils import np_utils
from sklearn import metrics 
```

```python
#model architecture
num_labels = y_test.shape[1]
def Model():
  model = Sequential()
  model.add(Conv2D(filters=16, kernel_size=2, input_shape=(70, 1320, 1), activation='relu'))
  model.add(MaxPooling2D(pool_size=2))
  

  model.add(Conv2D(filters=32, kernel_size=2))
  model.add(LeakyReLU(0.8))
  model.add(MaxPooling2D(pool_size=2))
  
  model.add(BatchNormalization(momentum=0.8))  
  model.add(Conv2D(filters=64, kernel_size=2))
  model.add(LeakyReLU(0.8))
  model.add(MaxPooling2D(pool_size=2))
  
  model.add(BatchNormalization(momentum=0.8))
  model.add(Conv2D(filters=128, kernel_size=2))
  model.add(LeakyReLU(0.8))
  model.add(MaxPooling2D(pool_size=2))
  
  model.add(GlobalAveragePooling2D())

  model.add(Dense(num_labels, activation='softmax'))

  return model
```

```python
model = Model()
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
# Display model architecture summary 
model.summary()

# Calculate pre-training accuracy 
score = model.evaluate(x_test, y_test, verbose=1)
accuracy = 100*score[1]

print("Pre-training accuracy: %.4f%%" % accuracy)
```

```python
from datetime import datetime
start = datetime.now()

history=model.fit(x_train , y_train , batch_size= 64 , epochs=100 ,verbose=1 , validation_data=(x_test , y_test) , shuffle=True)

duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
# history and loss


# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
def prediction(filename):
    mfcc_features = feature_extraction(filename)
    mfcc_features = mfcc_features.reshape(1 , 70, 1320, 1)
    predicted_vector = model.predict_classes(mfcc_features)
    
    predicted_class = le.inverse_transform(predicted_vector) 
    
    print("The predicted class is:", predicted_class[0], '\n') 

    predicted_proba_vector = model.predict_proba(mfcc_features) 
    
    predicted_proba = predicted_proba_vector[0]
    
    for i in range(len(predicted_proba)): 
        category = le.inverse_transform(np.array([i]))
        print(category[0], "\t\t : ", format(predicted_proba[i], '.32f') )
```
